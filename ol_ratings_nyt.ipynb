```python
# Load the CSV data into pandas DataFrames
ol_data = pd.read_csv(path + "title_ratings.csv", on_bad_lines='skip')
nyt_data = pd.read_csv(path + "books.csv", header=None)
```

```python
# Convert titles to lowercase and remove additional characters like parentheses
ol_data["cleaned_title"] = ol_data["title"].str.lower().str.replace(r'[\(\)]', '').str.strip()
nyt_data["cleaned_title"] = nyt_data[1].str.lower().str.replace(r'[\(\)]', '').str.strip()
```

```python
# Mark books in ol_data as New York Times bestsellers if their cleaned title is in nyt_data
ol_data["nyt_bestseller"] = ol_data["cleaned_title"].isin(nyt_data["cleaned_title"]).astype(int)
```

```python
# Create a new DataFrame with cleaned titles and nyt_bestseller column
new_df = ol_data[["cleaned_title", "average_rating", "nyt_bestseller"]]
```

```python
nyt_bestseller_df = new_df.loc[new_df["nyt_bestseller"] == 1]
```

```python
print(nyt_bestseller_df)
```

```python
# Count the occurrences of nyt_bestseller with value 1
nyt_bestseller_count = (new_df["nyt_bestseller"] == 1).sum()
print(nyt_bestseller_count)
```

```python

# Filter the DataFrame to show only rows where "nyt_bestseller" is equal to 1 (New York Times bestsellers)
bestseller_df = ol_data[ol_data["nyt_bestseller"] == 1]

# Count how many ratings of 4 and 5 for New York Times bestsellers
count_4 = (bestseller_df["average_rating"] == 4).sum()
count_5 = (bestseller_df["average_rating"] == 5).sum()

# Display the counts
print("Number of ratings 4 for New York Times bestsellers:", count_4)
print("Number of ratings 5 for New York Times bestsellers:", count_5)
```
